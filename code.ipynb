```python
import nltk
nltk.download('stopwords')
```

```python
import string
import re
from nltk.corpus import stopwords
stopwords=nltk.corpus.stopwords.words('english')
print(stopwords)
```

```python
import string
import re
from nltk.corpus import stopwords

def clean_text(text):
    text = "".join([char for char in text if char not in string.punctuation])
    tokens = re.split(r'\W+', text)  # Use raw string to avoid invalid escape sequence
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    text = [word for word in tokens if word not in stop_words]
    return text
```

```python
clean_text('I love NLP')
```

#TF-IDF VECTORIZER
APPLY TFIDVECTORIZER

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(analyzer=clean_text)
X_tfidf = tfidf_vect.fit_transform(messages['text'])
print(X_tfidf.shape)
print(tfidf_vect.get_feature_names_out())
```

```python
X_features=pd.DataFrame(X_tfidf.toarray())
X_features.head()
```

#using of random forest for classificaton

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_features,messages['label'],test_size=0.2)
X_train
```

```python
y_train
```

```python
rf = RandomForestClassifier()
rf_model = rf.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
```

```python
precision = precision_score(y_test, y_pred, pos_label='spam')
recall = recall_score(y_test, y_pred, pos_label='spam')
print('Precision: {} / Recall: {}'.format(round(precision, 3), round(recall, 3)))
```
